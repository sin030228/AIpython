```python
# while 문의 기본문법: while 조건식:
#while의 조건식이 참인 동안에만 반복된다.
# while True = 무한루프
```

```python
import random
answer = random.randint(1,100)
user = int(input("1부터 100사이의 정수 숫자를 맞추세요"))
count = 1
while user != answer:
  if user > answer:
     print("더 작은 숫자입니다")
     count +=1
  else:
     print("더 큰 숫자입니다")
     count+=1
  user = int(input("1부터 100사이의 정수 숫자를 맞추세요"))

print(f"정답입니다. {count}번만에 맞추셨습니다.")
```
