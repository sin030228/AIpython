```python
height1 = float(input("첫번째 학생의 키를 입력하세요(cm)"))
height2 = float(input("두번쨰 학생의 키를 입력하세요(cm)"))
hg = height1 - height2

if hg>=0:
  print(hg)

else:
  print(-hg)
```

```python
height1 = float(input("첫번째 학생의 키를 입력하세요(cm)"))
height2 = float(input("두번쨰 학생의 키를 입력하세요(cm)"))
if height1 > height2:
  print("첫번쨰 학생이 더 크다")
elif height1 < height2:
  print("두번쨰 학생이 더 크다")
else:
  print("두 학생의 키는 같다")
```

```python
import random

person1=random.randint(1,3)   # 1은 가위, 2는 바위 3은 보
person2=random.randint(1,3)

pg = person1-person2

print(f"첫번째사람{person1}")
print(f"두번째사람{person2}")

if person1> person2 and pg==1 :
  print("첫번째 사람이 이겼습니다")
elif person1> person2 and pg==2 :
  print("두번째 사람이 이겼습니다")
elif person1< person2 and pg==-1 :
  print("두번째 사람이 이겼습니다")
elif person1< person2 and pg==-2 :
  print("첫번째 사람이 이겼습니다")
else :
  print("비겼습니다")
```

```python
import random

person1=random.randint(1,3)   # 1은 가위, 2는 바위 3은 보
person2=random.randint(1,3)

pg = person1-person2

print(f"첫번째사람{person1}")
print(f"두번째사람{person2}")

if (person1> person2 and pg==1)or(person1< person2 and pg==-2) :
  print("첫번째 사람이 이겼습니다")
elif (person1> person2 and pg==2)or(person1< person2 and pg==-1) :
  print("두번째 사람이 이겼습니다")
else :
  print("비겼습니다")
```

```python
def hello(irum,age):
    oo = age+=1                 #이떄 hello 안에 irum은 매개변수라고 부른다. 매개변수는 여러개를 생성할 수 있다.
    print(f'{irum}님({age}세) 안녕하세요')
    return oo               # return 은 자신을 호출한 쪽으로 반환한다.
```

```python
hello('신동일',23)
```
