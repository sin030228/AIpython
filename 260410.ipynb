```python
i=0           # 시작값은 while 들어오기 전에
while i < 5:
  i+=1          # 증감값은 while 사이에
  if i == 3:
    continue
  print(i)      # 출력값은 while 끝난 후에


  # 증감값은 while 사이에 넣는 것이 일반적이지만 continue 문을 사용할때는 증감값의 위치를 주의해서 위치시켜야한다.
  # 리스트는 인덱스를 통해서 즉시 접근가능하다.
```

```python
joseon_kings = [
    "태조", "정종", "태종", "세종", "문종", "단종", "세조", "예종", "성종",
    "연산군", "중종", "인종", "명종", "선조", "광해군", "인조", "효종", "현종",
    "숙종", "경종", "영조", "정조", "순조", "헌종", "철종", "고종", "순종"
]

for i,k in enumerate(joseon_kings):
  print(f"{i+1}대 왕 {k}")

```

```python
number = int(input("몇대 왕이 누구인지 알고 싶은 번호를 적으세요"))

if 1 <= number <= len(joseon_kings):
  print(f"조선 {number}대 왕은 {joseon_kings[number-1]}입니다.")
```

```python
name = input("왕 이름을 입력하세요: ")

if name in joseon_kings:
  order = joseon_kings.index(name) +1         # 입력받은 이름이 "조선왕의 리스트에서 몇번째 위치인지"(인덱스) order라는 변수로 저장함
  print(f"{name}은 조선{order}대 왕입니다.")
```

```python
# in 연산자는  특정 데이터의 존재 여부를 알려준다.
# index() 함수는 특정 값의 위치를 반환한다. 실제위치는 인덱스 +1 로 표시
# joseon_kings.index(name) +1
# 조선왕 리스트/ name=입력받은 왕이름 의 인덱스/ +1 번째
joseon_kings.index("세종")
```
