```python
# 딕셔너리도 파이썬 리스트의 한 부류이다. -> 쌍으로 이루어진 데이터, 값에다 키를 다 부여하여 데이터로 저장할 수 있는 구조
# 문자열이나 리스트는 이름표가 없고 인덱스만 있다. 하지만 딕셔너리는 이름표(키)가 붙는다.
# 키를 찾으면 값이 나오는 구조이다. key & value
# 딕셔너리에서 사용하는 괄호는 중괄호이다.
# 형태: {key:value, key:value, .....}
# 리스트는 인덱스를 통해서 접근하지만 딕셔너리는 key를 통해서 값에 접근하는 방법이다.
hi = { '1':'me', '2':'you'}
```

```python
cafe_menu = {
    "아메리카노": 4500,
    "카페라떼": 5200,
    "바닐라라떼": 5700,
    "카푸치노": 5300,
    "카라멜 마키아토": 5900,
    "콜드브루": 5600,
    "카페모카": 6000,
    "초코라떼": 5500,
    "말차라떼": 5800,
    "레몬에이드": 6200
}

total = 0

while True:
    menu = input("메뉴를 입력하세요(종료 입력 시 끝): ").strip()

    if menu == "종료":
        break

    if menu in cafe_menu:
        total += cafe_menu[menu]
        print(f"추가됨: {menu} ({cafe_menu[menu]}원)")
    else:
        print("없는 메뉴입니다. 다시 입력해 주세요.")

print(f"최종 합계: {total}원")
```
