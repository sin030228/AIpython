```python
# 아나콘다 Anaconda Navigator
# Visual Studio Code: 다양한 언어들을 편집하고 실행하는 프로그램 C, 파이썬, Java 등 전세계 개발자들이 많이 사용하고 있다.
# 구글 코렙은 ipynb 확장자를 사용하고 파이썬은 py 확장자를 사용한다.
# 식별자와 키워드의 구분 - 키워드는 이미 프로그램 내에서 약속된 단어: 예약어, 따라서 변수 = 식별자
# "=" 대입연산자의 기준으로 왼쪽에는 변수, 오른쪽에는 값 또는 변수가 온다. 이떄 오른쪽에 오는 값을 P연산자라고 사용한다. 연산자의 성격은 P연산자의 성격에 의해 정해진다.
# 대입연산자의 오른쪽 부터 계산을 시작한다.
#변수는 한번에 하나의 값만 저장한다. 일반적으로 변수의 이전 값은 지워진다.
# 변수로 저장할 수 있는 종류 = 자료형 (Data Type)
# 어떤 변수가 어떤 자료형을 가지고 있는지 알아보려면 "type()" 함수를 사용하면 알 수 있다.
# 자료형의 종류 1. 정수형(int) 2. 부동소수점형(float) 3. 문자열형(str) 4. 불형(bool)
# 정수형은 integer의 줄임말, float는 floating point number의 줄임말, str은 string의 줄임말, bool은 boolean의 줄임말 True False <- 대소문자 구분!
# +는 원래 산술 연산자 이지만, 문자열을 연결할때 +를 사용하면 문자열들을 이어붙일 수 있다. 이때 +는 문자열 연결 연산자로 사용된 것이다.
```

```python
# 변수 명명 규칙
# 1. 첫글자는 반드시 문자 또는 밑줄로 시작해야한다. 숫자로 시작할 수 없다.
# 2. 변수의 구성요소는 문자, 숫자, 밑줄로만 구성되어야 한다. 특수문자 또는 공백은 사용할 수 없다.
# 3. 대소문자를 구분하여 변수를 생성해야한다.
# 4. 키워드(예약어) 는 파이썬 언어자체에서 이미 용도를 정해둔 단어이기 때문에 사용할 수 없다. ex) if, print
```

```python
# 변수의 값 교환하기
# a = 7, b = 11일 때 두 변수에 저장된 값을 바꾸기 위해서는 c라는 변수를 하나 만든다.
# c = 0, c = a, a = b, b = c 로 변수를 지정하면 된다. c = 7, a = 11, b = 7
# 일반적인 프로그래밍 언어라면 위의 예시로 변수의 값을 바꿔야 하지만 파이썬에서는 간단하게 바꿀 수 있다.
# a, b = b, a
# 위의 식을 사용하면 a = 11, b = 7의 값으로 바뀌게 된다. 이 문법의 핵심은 대입 연산자(=)의 오른쪽(b, a)이 항상 먼저 계산(평가)된다는 점이다.
```

```python
#1) 산술 연산자 (Arithmetic)
#+ 덧셈
#- 뺄셈
#* 곱셈
#/ 나눗셈(항상 실수 결과 가능)
#// 몫(내림된 정수 몫)
#% 나머지
#** 거듭제곱
#예: 7 // 2 == 3, 7 % 2 == 1, 2 ** 3 == 8
```

```python
a = int(input('지금은 몇분입니까?'))
h = a//60
m = a%60
print('총 시간은', h, '시간', m, '분 입니다.')
```

```python
# 4) 할당 연산자 (Assignment)
#= 기본 할당
#복합 할당: +=, -=, *=, /=, //=, %=, **=
#비트 복합 할당: &=, |=, ^=, <<=, >>=
#대입 표현식(바다코끼리 연산자): := (표현식 안에서 값 할당)
#예: x += 1 는 x = x + 1
```

```python
minutes = int(input("숫자(분 단위)를 입력하세요: "))
hours = minutes // 60
mins = minutes % 60

print(f"{minutes}분은 {hours}시간 {mins}분입니다.")
```

```python
# 1) 사용자에게 숫자를 입력받는다 (input은 문자열이므로 int로 바꿔줌)
num = int(input("숫자를 입력하세요: "))

# 2) 2로 나눈 나머지(%)를 이용해서 짝수/홀수 판단
#    - 나머지가 0이면 짝수
#    - 나머지가 1이면 홀수
if num > 0:
    print(f"{num}은(는) 양수입니다.")
else:
    print(f"{num}은(는) 음수입니다.")
```
