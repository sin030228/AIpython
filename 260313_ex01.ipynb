```python
# 1. 입력 받기 (사용자가 키보드로 숫자를 넣음)
# input() 함수는 사용자가 입력한 값을 가져옵니다.
# 하지만 input은 무조건 '문자(글자)'로 가져오기 때문에,
# int()를 감싸서 '정수(숫자)'로 바꿔줘야 계산이 됩니다.

s_num1 = input("첫 번째 숫자를 입력하세요: ") # 입력을 받아서
num1 = int(s_num1)                         # 숫자로 바꿔서 num1에 저장

s_num2 = input("두 번째 숫자를 입력하세요: ") # 입력을 받아서
num2 = int(s_num2)                         # 숫자로 바꿔서 num2에 저장

# 2. 연산 수행 (변수에 저장된 값으로 계산하기)
# (여기는 기존 코드와 완전히 똑같습니다!)

# 덧셈 연산자 (+)
result_add = num1 + num2

# 뺄셈 연산자 (-)
result_sub = num1 - num2

# 곱셈 연산자 (*)
result_mul = num1 * num2

# 나눗셈 연산자 (/)
result_div = num1 / num2

# 3. 결과 출력 (계산된 값을 화면에 보여주기)
print("--------------------")
print("입력한 첫 번째 숫자:", num1)
print("입력한 두 번째 숫자:", num2)
print("--------------------")
print("더하기 결과:", result_add)
print("빼기 결과:", result_sub)
print("곱하기 결과:", result_mul)
print("나누기 결과:", result_div)
```
